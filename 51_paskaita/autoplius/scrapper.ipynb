```python
from selenium import webdriver
import os
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import requests

options = Options()
options.add_argument('--incognito')

driver = webdriver.Chrome(options=options)

url = 'https://m.autoplius.lt/skelbimai/naudoti-automobiliai'

driver.get(url)

photos = driver.find_elements(By.CLASS_NAME, 'js-gallery-photo')
```

```python
for i, photo in enumerate(photos):
    print(i, photo.get_attribute('src'))
```

```python

photos_dir = 'automobiliai'
if not os.path.exists(photos_dir):
    os.makedirs(photos_dir)

for i, photo in enumerate(photos):
    photo_url = photo.get_attribute('src')
    img_data = requests.get(photo_url).content
    img_name = f'{i}.jpg'
    img_path = os.path.join(photos_dir, img_name)
    with open(img_path, 'wb') as h:
        h.write(img_data)
    print(f'Issaugota {i} nuotrauka')
driver.quit()
```

```python
# traktoriai
from selenium import webdriver
import os
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import requests

options = Options()
options.add_argument('--incognito')

driver = webdriver.Chrome(options=options)
import time
from random import randint
options = Options()
options.add_argument('--incognito')

driver = webdriver.Chrome(options=options)

for page_nr in range(1,100):

    url = f'https://m.autoplius.lt/skelbimai/zemes-ukio-misko-technika-manipuliatoriai/zemes-ukio-savaeige-technika?model=&qt=&category_id=32&type_id%5B6434%5D=6434&page_nr={page_nr}'

    driver.get(url)

    photos = driver.find_elements(By.CLASS_NAME, 'not-lazy')

    for i, photo in enumerate(photos):
        print(i, photo.get_attribute('src'))

    photos_dir = 'traktoriai'
    if not os.path.exists(photos_dir):
        os.makedirs(photos_dir)

    for i, photo in enumerate(photos):
        photo_url = photo.get_attribute('src')
        img_data = requests.get(photo_url).content
        img_name = f'{page_nr}_{i}.jpg'
        img_path = os.path.join(photos_dir, img_name)
        with open(img_path, 'wb') as h:
            h.write(img_data)
        print(f'Issaugota {i} nuotrauka')

    time.sleep(randint(1,5))
driver.quit()
```
